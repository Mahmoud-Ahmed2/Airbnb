```python
sending = df.groupby('dim_device_app_combo').sum()
sending.sort_values("sent_message",ascending= False, inplace= True)
sending
```

```python
sending.plot.bar(y='sent_message')
```

```python
df['ts_min']=pd.to_datetime(df['ts_min'])
df['ts_max']=pd.to_datetime(df['ts_max'])
```

```python
df['time_spend'] = df['ts_max'] - df['ts_min']
df.columns
```

```python
# group by device-app combo
time_spend = df.groupby('dim_device_app_combo')['time_spend'].mean().sort_values()

# Plotting the bar chart
time_spend.plot(kind='bar', figsize=(10,6), legend=False)
plt.ylabel('Average Time Spent')
plt.title('Average Time Spent by Device-App Combo')
plt.show()
```
